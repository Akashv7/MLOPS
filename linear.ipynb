```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.select_dtypes(include=['object']).columns:
    df[col] = le.fit_transform(df[col])

df.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

print("="*60)
print("LINEAR REGRESSION: CRIME FREQUENCY PREDICTION")
print("="*60)

# Create a continuous target variable for linear regression
# Predicting crime frequency/count by district
df['Crime_Frequency'] = df.groupby('District')['District'].transform('count')

print("\nTarget Variable: Crime_Frequency (Continuous)")
print(f"Target Statistics:")
print(df['Crime_Frequency'].describe())

# Prepare features (X) and target (y) for linear regression
X = df.drop(columns=['Crime_Frequency', 'Unnamed: 0', 'ID', 'Case Number', 'Date', 'Updated On', 'Location', 'Primary Type', 'Description', 'Location Description'])
y = df['Crime_Frequency']

print(f"\nFeatures shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Features being used: {list(X.columns)[:5]}... (and {len(X.columns)-5} more)")
```

```python
# Train-Test Split for Linear Regression
from sklearn.preprocessing import PolynomialFeatures

# Use 90-10 split to give model more training data for better fit
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Create polynomial features (degree 3) 
poly = PolynomialFeatures(degree=3, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_poly)
X_test_scaled = scaler.transform(X_test_poly)

print("\n✓ Data Split Complete with Polynomial Features:")
print(f"  Original features: {X.shape[1]}")
print(f"  Polynomial features (degree 3): {X_train_scaled.shape[1]}")
print(f"  Training set size: {X_train_scaled.shape}")
print(f"  Testing set size: {X_test_scaled.shape}")
```

```python
# Train Linear Regression Model with Zero Regularization
import warnings
warnings.filterwarnings('ignore')

print("\nTraining Linear Regression Model with Polynomial Features (Degree 3)...")

# Use standard LinearRegression with no regularization
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred_train = model.predict(X_train_scaled)
y_pred_test = model.predict(X_test_scaled)

print("✓ Linear Regression Model trained successfully!")
print(f"\nModel Parameters:")
print(f"  Coefficients shape: {model.coef_.shape}")
print(f"  Model Intercept: {model.intercept_:.4f}")
print(f"  Model Type: Standard LinearRegression (Zero Regularization)")

# Get feature importance
top_indices = np.argsort(np.abs(model.coef_))[-5:][::-1]
print(f"\nTop 5 Most Important Features (Polynomial):")
for i, idx in enumerate(top_indices):
    print(f"  {i+1}. Feature {idx}: {model.coef_[idx]:.4f}")
```

```python
# Evaluate Linear Regression Model
print("\n" + "="*60)
print("LINEAR REGRESSION MODEL EVALUATION")
print("="*60)

# Calculate regression metrics
train_mse = mean_squared_error(y_train, y_pred_train)
test_mse = mean_squared_error(y_test, y_pred_test)
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)
train_mae = mean_absolute_error(y_train, y_pred_train)
test_mae = mean_absolute_error(y_test, y_pred_test)
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

print("\nRegression Metrics:")
print(f"Train R² Score:        {train_r2:.4f}")
print(f"Test R² Score:         {test_r2:.4f}")
print(f"\nTrain MSE:             {train_mse:.4f}")
print(f"Test MSE:              {test_mse:.4f}")
print(f"\nTrain RMSE:            {train_rmse:.4f}")
print(f"Test RMSE:             {test_rmse:.4f}")
print(f"\nTrain MAE:             {train_mae:.4f}")
print(f"Test MAE:              {test_mae:.4f}")

print("\n" + "="*60)
print("SAMPLE PREDICTIONS (Crime Frequency by District):")
print("="*60)
for i in range(min(5, len(y_test))):
    actual = y_test.iloc[i]
    predicted = y_pred_test[i]
    error = abs(actual - predicted)
    print(f"Sample {i+1}: Actual={actual:.0f}, Predicted={predicted:.2f}, Error={error:.2f}")
```
