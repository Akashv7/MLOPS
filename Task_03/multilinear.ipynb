```python


import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv('hiring.csv')
word_to_num = {
    'one':1, 'two':2, 'three':3, 'four':4, 'five':5,
    'six':6, 'seven':7, 'eight':8, 'nine':9,
    'ten':10, 'eleven':11, 'twelve':12
}
df['experience'] = df['experience'].map(word_to_num)
df['experience'] = df['experience'].fillna(0)          
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].mode()[0]              
)
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']
model = LinearRegression()
model.fit(X, y)
print("R² score:", model.score(X, y))   
p1 = model.predict([[2, 9, 6]])
p2 = model.predict([[12, 10, 10]])
print(f"Candidate 1: ${p1[0]:.2f}")    
print(f"Candidate 2: ${p2[0]:.2f}")
```
